```python
import os
import numpy as np
from glob import *
from scipy.integrate import *
import mesa_reader as mr
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
path = 'C:/Users/seayc/Documents/school/masters/classes/year_one/sse/assignment/sse/2M_prems_to_wd/LOGS/'
path
```

```python
# hr diagram
h = mr.MesaData(path+'history.data')
plt.plot(h.log_Teff[0:200],h.log_L[0:200],c='k') # pre-main sequence
plt.plot(h.log_Teff[200:300],h.log_L[200:300],c='c') # main sequence
plt.plot(h.log_Teff[300:5000],h.log_L[300:5000],c='r') # red-giant branch (agb) phase
plt.plot(h.log_Teff[5000:5865],h.log_L[5000:5865],c='b') # white dwarf ascension

# annotations
# plt.ann

# aesthetics 
plt.xlabel('log T$_{\mathrm{eff}}$ [K]',fontsize = 13)
plt.ylabel(r'log L [L/L$_{\odot}$]',fontsize = 13)
plt.gca().invert_xaxis()
```

```python
# temperature vs density

# load entire LOGS directory information
l = mr.MesaLogDir(path)

# grab the last profile
p = l.profile_data()

# this works even if you only have logRho and logT!
plt.plot(h.log_center_T, h.log_center_Rho, c='k')
plt.xlabel(r'log T$_c$ [K]', fontsize = 13)
plt.ylabel(r'log $\rho_c$ [g cm$^{-3}$]', fontsize = 13)
```

```python
# convection in the pre-main sequence
pre_ms = l.profile_data(profile_number=4) # pre-main sequence
x = pre_ms.star_age
print(f'profile is {round(x/1e9,5)} billion years old')
plt.semilogy(pre_ms.R,pre_ms.grada,c='r')
plt.semilogy(pre_ms.R,pre_ms.gradr,c='b')

plt.xlabel(r'R [R/R$_\odot$]',fontsize = 13)
plt.ylabel(r'$\nabla T$ [K]',fontsize = 13)
```

```python
# convection in the main sequence
ms = l.profile_data(profile_number=9) # main sequence
y = ms.star_age
print(f'profile is {round(y/1e9,6)} billion years old')
plt.semilogy(ms.R,ms.grada,c='r')
plt.semilogy(ms.R,ms.gradr,c='b')

plt.xlabel(r'R [R/R$_\odot$]', fontsize = 13)
plt.ylabel(r'$\nabla T$ [K]', fontsize = 13)
```
